### 1. Scatter Plot

#### i} Plotting the Scatter Plots

#### ii} Labeling the data points (co-ordinates)

```python
x1 = 3 * np.random.rand(1000, 2)
y1 = 4 + 3 *x1* np.random.randn(1000, 2)

x2 = 2 * np.random.rand(1000, 2)
y2 = 2 + 4 *x1* np.random.randn(1000, 2)


dataset = [(x1, y1), 
           (x2, y2)]


plt.scatter(dataset[0][0], dataset[0][1], c="red", s=20, label="From X1, Y1")
plt.scatter(dataset[1][0], dataset[1][1], c="green", s=100, label="From X2, Y2")
plt.title("Labelled data")
plt.xlabel("$X axis$")
plt.ylabel("$Y axis$")
# For enabling the labeling
plt.legend()
plt.show()
```

#### iii} Changing the shapes of "dots" (Using markers)

```python
plt.scatter(dataset[0][0], dataset[0][1], c="red", s=20, label="From X1, Y1", marker="*")
plt.scatter(dataset[1][0], dataset[1][1], c="green", s=100, label="From X2, Y2", marker="^")
plt.title("Labelled data")
plt.xlabel("$X axis$")
plt.ylabel("$Y axis$")
plt.legend()
plt.show()
```

#### iv} Changing the edge colors

```python
plt.scatter(dataset[0][0], dataset[0][1], c="red", s=100, label="From X1, Y1", marker="o", edgecolors="k")
plt.scatter(dataset[1][0], dataset[1][1], c="green", s=100, label="From X2, Y2", marker="^", edgecolors="y")
plt.title("Labelled data")
plt.xlabel("$X axis$")
plt.ylabel("$Y axis$")
plt.legend()
plt.show()
```

#### v} Shorthand syntax for scatter plots

```python
plt.plot(dataset[0][0], dataset[0][1], "ro", label="From X1, Y1")
plt.plot(dataset[1][0], dataset[1][1], "g^", label="From X2, Y2")
plt.title("Labelled data")
plt.xlabel("$X axis$")
plt.ylabel("$Y axis$")
plt.legend()
plt.show()
```

#### vii} Sub-plots

```python
dataset = [(x1, y1), 
           (x2, y2)]

plt.subplot(1, 2, 1)
plt.plot(dataset[0][0], dataset[0][1], c="red",label="Red Line")
plt.title("Red line")
plt.xlabel("$X-axis$")
plt.ylabel("$Y-axis$")
plt.legend()


plt.subplot(1, 2, 2)
plt.plot(dataset[1][0], dataset[1][1], c="green", label="Green Line")
plt.title("Green line")
plt.xlabel("$X-axis$")
plt.ylabel("$Y-axis$")
plt.legend()

plt.subplots_adjust(left=2, right=4)
plt.show()
```

##### c} Stacked Bar Graph

####o Vertical 

```python
X = ["USA", "China", "Russia"]

gold = np.array([20, 10, 7])
bronze = np.array([31, 21, 20])
silver = np.array([41, 28, 19])

plt.bar(X, gold,  color="y", label="gold")
plt.bar(X, bronze, bottom=gold, color="r", label="bronze")
plt.bar(X, silver, bottom=gold + bronze, color="g", label="silver")
plt.legend()
plt.show()
```

#### o Horizontal

```python
X = ["USA", "China", "Russia"]

gold = np.array([20, 10, 7])
bronze = np.array([31, 21, 20])
silver = np.array([41, 28, 19])

plt.barh(X, gold,  color="y", label="gold")
plt.barh(X, bronze, left=gold, color="r", label="bronze")
plt.barh(X, silver, left=gold + bronze, color="g", label="silver")
plt.legend()
plt.show()
```

### 4. Pie Charts

#### i} Plotting Chart

```python
omlete_ingredient = ["Egg", "Oil", "Spices", "Stove"]
percetage_ing = np.array([35, 20, 15, 30])

# The figsize attribute allows us to specify the width and height of a figure in unit inches.

plt.figure(figsize=(16, 8))

plt.pie(percetage_ing, labels=omlete_ingredient)
plt.title("Omlete Ingredients")
plt.show()
```

#### ii} Exploding the wedges

```python
exploding_values = [0.2, 0, 0, 0]

omlete_ingredient = ["Egg", "Oil", "Spices", "Stove"]
percetage_ing = np.array([35, 20, 30, 15])

# The figsize attribute allows us to specify the width and height of a figure in unit inches.

plt.figure(figsize=(16, 8))

plt.pie(percetage_ing, labels=omlete_ingredient, explode=exploding_values)
plt.title("Omlete Ingredients")
plt.show()
```

#### iii} Changing the colors

```python
omlete_ingredient = ["Egg", "Oil", "Spices", "Stove"]
percetage_ing = np.array([35, 20, 30, 15])
colors = ["gray", "yellow", "red", "black"]


# The figsize attribute allows us to specify the width and height of a figure in unit inches.

plt.figure(figsize=(16, 8))

plt.pie(percetage_ing, labels=omlete_ingredient, colors=colors)
plt.title("Omlete Ingredients")
plt.legend()
plt.show()
```

#### iv} Showing percentage with *autopct* paramter

```python
omlete_ingredient = ["Egg", "Oil", "Spices", "Stove", "Pepper"]
percetage_ing = np.array([35, 20, 30, 10, 5])
colors = ["gray", "yellow", "red", "green", "gold"]


# The figsize attribute allows us to specify the width and height of a figure in unit inches.

plt.figure(figsize=(16, 8))

plt.pie(percetage_ing, labels=omlete_ingredient, colors=colors, autopct="%1.1f%%")
plt.title("Omlete Ingredients")
plt.legend()
plt.show()
```
